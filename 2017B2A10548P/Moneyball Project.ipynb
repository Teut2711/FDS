<h2>Feature Engineering:    <i>Batting</i></h2>
<ul>
   <li> Batting Average :<br>
        \begin{align}    
        BA = \frac{H}{AB} 
        \end{align}
   </li><br> 
   <li> On Base Percentage :
       <br> <br>
        \begin{align}    
        OBP = \frac{H + BB + HBP}{AB + BB + HBP +SF } 
        \end{align}
   </li><br>    
   <li> Slugging Percentage : 
       <br><br>
       \begin{align}    
        SP = \frac{1 B + (2 \times 2B) + (3 \times 3B) +(4 \times HR)}{AB} 
        \end{align}
   </li>    
   <br><br>
    <li> Runs per game : 
       <br><br>
       \begin{align}    
        AR = \frac{R}{G} 
        \end{align}
   </li>    
   <br><br>
   \begin{align}    
    H = \ B + 2B + 3B + HR  
   \end{align}


   


```r
batting <- na.omit(batting, cols=c(AB:HR,"HBP","SF"))
batting <- batting %>% mutate(BA=H/AB, OBP=(H+BB+HBP)/(AB+BB+HBP+SF), B=H-(X2B+X3B+HR), SP=(B+2*X2B+3*X3B+4*HR)/AB, score=(1+.7*.7)*(OBP*SP)/(0.7*.7*OBP+SP), AR=R/G)
```

```r
batting <- batting %>% filter(!is.nan(score), !is.nan(AR)) 
```

```r
options(repr.matrix.max.rows=nrow(batting), repr.matrix.max.cols=ncol(batting))

head(batting)
```

<u><h1>How to choose players ?- <span style="color:yellow;font-style:italic;font-family: 'Dokdo', cursive;">Algorithm</span></h1></u>

<p><i>We iterate through the rows picking 3 rows at a time and then multiplying number of games offered to the three lost players to corresponding players in batters_pool(the three rows) by multiplying max games to player with highest OBP,and then SP since OBP determines the ability of the player to remain at base and stay longer in game.  </i></p>

```r
players_to_choose <- data.frame(Player1=" ", Player2=" ", Player3=" ",R=0,Total_spendings=0 ,stringsAsFactors = F) 
i = 1

for (row1 in 1:nrow(tail(batters_pool,50))){
 for (row2 in row1:nrow(tail(batters_pool,50)))
    for (row3 in row2:nrow(tail(batters_pool,20))){
      temp <- rbind(batters_pool[row1,],batters_pool[row2,],batters_pool[row3,]) 
      temp<- temp %>% arrange(OBP, SP) 
      for (k in 1:3){
        if (temp[k,"AR"] >= lost_players[k,"AR"]){
           temp[k,"R"] <- temp[k,"AR"]*lost_players[k,"G"]}  }
      if ((batters_pool[row1,]$salary + batters_pool[row2,]$salary + batters_pool[row3,]$salary <= money_we_have)&
        (batters_pool[row1,]$R + batters_pool[row2,]$R + batters_pool[row3,]$R >=250))          {
         players_to_choose<- rbind(players_to_choose,c(as.character(batters_pool[row1,]$playerID) ,as.character(batters_pool[row2,]$playerID), as.character(batters_pool[row3,]$playerID), batters_pool[row1,]$R + batters_pool[row2,]$R + batters_pool[row3,]$R, batters_pool[row1,]$salary + batters_pool[row2,]$salary + batters_pool[row3,]$salary)) 
         i <- i + 1       
}}}
```

<p>Predicting whether a team would join class 0 or not joining OAK's or 1  join OAK's. For this I take the data from players pool and assign 1 if he belongs to "OAK" and 0 otherwise </p>

```r
batters_2002 <- batting20 %>%filter(teamID =="OAK", yearID==2002) #players who are in batters list of 2002 and in OAK
batters_2002 <- batters_2002 %>% filter(!(playerID %in% team_without$playerID)) 
test <- batting20 %>% filter(yearID==2002) %>% mutate(label=ifelse(playerID %in% batters_2002$playerID,1,0)) #test data
```

```r
train <- batting19 %>% mutate(label=ifelse(teamID=="OAK",1,0)) # assign class 1 to in OAK and 0 to not in OAK
logistic <- glm( label~ OBP + SP + AR  + BA , data = train,family = binomial)
```

```r
summary(logistic) 
```

```r
li.null <-logistic$null.deviance/-2
li.proposed <-logistic$deviance/-2
(li.null-li.proposed)/li.null
```

#### By above result  p value <.1 and thus null hypothesis is quite significant.Too bad for the model...

```r
x <- predict(logistic, test, type="response") 
```

```r
dt <- test %>% mutate(predicted = ifelse(x>  0.5,1, 0))   #alpha- value here gives high misclassification
x_ <- dt %>% filter(label != predicted) %>% count()
```

```r
x_ <- dt %>% filter(label != predicted) %>% count()
x_  #Incorrect Classification
```
